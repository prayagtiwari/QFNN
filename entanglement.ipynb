```python
import qiskit
import qleet
tc=qiskit.QuantumCircuit(6)
params= [qiskit.circuit.Parameter(f"θ{i}") for i in range(12)]
for i in range(6):
    tc.h(i)
    tc.rz(params[i], i)
tc.cnot(0,1)
tc.cnot(1,2)
tc.cnot(2,0)
tc.cnot(3,4)
tc.cnot(4,5)
tc.cnot(5,3)

for i in range(6):
    tc.ry(params[i+6], i)

tc.draw("mpl")
```

```python
qiskit_descriptor = qleet.interface.circuit.CircuitDescriptor(
    circuit=tc, params=params, cost_function=None
)
qiskit_entanglement_capability = (
    qleet.analyzers.entanglement.EntanglementCapability(
        qiskit_descriptor, samples=1000
    )
)
entanglement_mw = qiskit_entanglement_capability.entanglement_capability("meyer-wallach")
entanglement_mw
```
